# DELTA X AND Y

```python
df_info[['Deltax', 'Deltay']].plot(linestyle='-', marker='o', figsize=(10,6), title = 'DeltaX and DeltaY per Policy Execution (50 runs) of $\pi_{imitated_{1}}$', xlabel = 'Policy Executions', ylabel= 'Values')


plt.show()
```

```python
df_deltassamuel = pd.read_csv('Samuel/data.csv')
df_deltassamuel
```

```python
df_deltassamuel['delta_x'].values
```

```python



j = 0
deltax = []
deltay = []
for i in df_infosamuel['Samples'].values:
    deltax.append(sum(df_deltassamuel['delta_x'][j:i - 1])/i)
    deltay.append(sum(df_deltassamuel['delta_y'][j:i - 1])/i)
    j = i
     
```

```python
deltas = pd.DataFrame(deltax, columns=['Deltax'])
deltas

deltay = pd.DataFrame(deltay, columns=['Deltay'])
deltay
```

```python
deltas = pd.concat([deltas, deltay], axis=1)
deltas.head()
```

```python
deltas.plot(linestyle='-', marker='o', figsize=(10,6), title = 'DeltaX and DeltaY per Policy Execution (50 runs) of $\pi_{referential}$', xlabel = 'Policy Executions', ylabel= 'Values')
plt.show()
```
