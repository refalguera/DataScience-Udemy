### Matplotlib

Matplotlib is the most popular plotting library for python. It gives you control every aspect of a figure.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
```

```python
##Matplotlib objetc oriented programming
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8]) ##left, bottom, width and higth of the axes
axes.plot(x,y)
axes.set_xlabel('Xlabel')
axes.set_ylabel('Ylabel')
axes.set_title('Title')
```

#### Figure Size and DPI

```python
fig = plt.figure(figsize=(8,2),dpi=100) ##dpi pixels per inch

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2, ncols=1,figsize=(8,2),dpi=100) ##dpi pixels per inch

axes[0].plot(x,y)

axes[1].plot(y,x)

plt.tight_layout()
```

```python
##LEGENDS
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])


axes1.plot(x,x**2, label='Function1')
axes1.plot(x,x**4, label='Function 2')

axes1.legend()

axes1.set_title('Title')
axes1.set_xlabel('X')
axes1.set_ylabel('Y')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])


axes1.plot(x,x**2, label='Function1')
axes1.plot(x,x**4, label='Function 2')

##best location for the legend --> use loc
axes1.legend(loc=0)

axes1.set_title('Title')
axes1.set_xlabel('X')
axes1.set_ylabel('Y')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])


axes1.plot(x,x**2, label='Function1')
axes1.plot(x,x**4, label='Function 2')

##best location for the legend --> use loc=0
axes1.legend(loc=10)

axes1.set_title('Title')
axes1.set_xlabel('X')
axes1.set_ylabel('Y')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])


axes1.plot(x,x**2, label='Function1')
axes1.plot(x,x**4, label='Function 2')

##best location for the legend --> use loc=0
axes1.legend(loc=(0.1,0.1))

axes1.set_title('Title')
axes1.set_xlabel('X')
axes1.set_ylabel('Y')
```
